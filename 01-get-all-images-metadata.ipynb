```python
import json
import pandas as pd
import urllib.request
```

```python
def get_url(x):
    contents = urllib.request.urlopen(x).read()
    return json.loads(contents.decode("utf-8"))
```

```python
def parse_one_result(x):
    return {'timestamp': x['timestamp'],
            'lat': x['geometrie']['coordinates'][0],
            'lon': x['geometrie']['coordinates'][1],
            'href': x['_links']['self']['href']
           }
```

```python
def parse_all_results_in_one_page(x):
    return pd.DataFrame.from_records([parse_one_result(y) for y in x])
```

```python
all_results = []
```

```python
one_page = get_url("https://api.data.amsterdam.nl/panorama/recente_opnames/2017/?page_size=1000")
this_results = parse_all_results_in_one_page(one_page['results'])
all_results.append(this_results)
while True:
    next_page = one_page['_links']['next']['href']
    print(next_page)
    one_page = get_url(next_page)
    this_results = parse_all_results_in_one_page(one_page['results'])
    all_results.append(this_results)
```
